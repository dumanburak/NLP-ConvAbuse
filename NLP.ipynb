# Cleaning Data with Regular Expression

```python
def cleanTxt(text):
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'_', '', text)
    text = re.sub(r'^\s*', '', text)
    text = re.sub(r'[0-9]', '', text)
    return text
```

#  Removing Stop Words

```python
def list_to_string(s): # list
    str1 = ""
    
    # traverse in the string
    for ele in s:
        str1 += ele +" "
    # return string
    return str1
```

```python
sentence_without_stopwords = []

stop_words = set(stopwords.words('english'))

for i in range(len(sentence)):
    word_tokens = word_tokenize(sentence_list[i])

    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

    filtered_sentence = []

    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)

    #print(word_tokens)
    sentence_without_stopwords.append(list_to_string(filtered_sentence))
    #print(i+1, ":", list_to_string(filtered_sentence))
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

# Spliting our data train and test values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) # 25% test, 75% train
```

# Normalization â€“ Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

<h1><span style="color:blue">Creating Gaussian Naive Bayes Model and Training</span></h1>

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

<h3><span style="color:red">Confusion Matrix and Accuracy Score of Gaussian Naive Bayes Model</span></h3> 

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: \n\n",cm)
```

```python
# printing the accuracy of the model
print("Accuracy Score is:",accuracy_score(y_pred, y_test))
```

<h1><span style="color:blue">Creating Bernoulli Naive Bayes Model and Training</span></h1>

```python
from sklearn.naive_bayes import BernoulliNB

# initializaing the NB
classifer = BernoulliNB()

# training the model
classifer.fit(X_train, y_train)

# testing the model
y_pred = classifer.predict(X_test)
```

<h3><span style="color:red">Confusion Matrix and Accuracy Score of Bernoulli Naive Bayes Model</span></h3>

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: \n\n",cm)
```

```python
# printing the accuracy of the model
print("Accuracy Score is:",accuracy_score(y_pred, y_test))
```
